We will start of the Phase Two of the EDA by interpreting the result of the correlation matrix we previously plotted. Per our correlation matrix above, the highest correlation is between our 'left' variables which indicates whether the employee left the company (1 if yes, 0 if no) and satisfaction level of the employee. 

There is a moderate negative correlation between these two variables, -0.35, which means that as the satisfaction level increases, the employee is less likely to leave, and as the satisfaction level decreases, the employee is more likely to leave. So now as the first step let's look further into the satisfaction level. 

Our goal is to look into the correlations, counts and means at a departmental level to identify which departments have the employees with the lowest satisfaction level and highest likelihood to leave company. Our hypothesis is that employees who had work accidents often (work accident), got minimum to no promotions in the past 5 years (promotions in last 5 years) and are in the low pay group (salary) are likely to leave. 

This study aims to clearly present the correlation between the above variables and employees departure, which will give HR and the leadership of the company a specific group and certain problematic conditions to focus in order to minimize employee turnover.

```python
working_df = df[['satisfaction_level','department','work_accident','salary','promotion_last_5years','left']]
working_df.head()
```

Now let's only keep the departments mention in the above paragraph, that are 'HR','Accounting' and 'Technical'

```python
working_df = working_df[working_df["department"].isin(["hr", "accounting","technical"])]
working_df.describe()
```

```python
working_df.groupby('department').count()
```

```python
sns.countplot(x="department", data=working_df)
```

The largest department we will be further investigating is 'technical' with 2,244 past and present employees followed by accounting and hr with 621 and 601, respectively.

```python
working_df.groupby('department').mean()
```

Looking at the overall mean and median of 'satisfaction level', 'average monthly hours', 'work accidents', 'promotions in last 5 years, we see that even though 'technical' department have the highest number of monthly average work hours and work accidents together with lowest number of promotions, they have the highest satisfaction level and lowest departure rate. 

Accounting department employees have the lowest satisfaction level while their overall mean values are in the mid segment, less than 'technical' department and more than 'hr' in general.

We see that HR department employees despite being the group that works the least average monthly hours, has the least work accident rate and highest promotion rate are also the employees that have the highest departure rate. 

Now let's look at the salary group distribution for these departments.

```python
dept_salary = pd.DataFrame(working_df.groupby(["department", "salary"])["salary"].count().reset_index(name="count"))
dept_salary.style.background_gradient(cmap = 'Reds')
```

```python
sns.countplot(x="department", hue="salary", data=working_df)
```

```python
dept_salary_perc = dept_salary.groupby(['department', 'salary']).agg({'count': 'sum'})
dept_salary_perc.groupby(level=0).apply(lambda x:100 * x / float(x.sum()))
```

Accounting department have the highest percentage of employees that are in the high salary group also the lowest percentage of employees in the low salary group. This results gives us an interesting insight as Accounting department employees also have the lowest satisfaction rate compared to HR and Technical department employees. 

Now we will further look into the indicators of departure within these departments

```python
salary_left = pd.DataFrame(working_df.groupby(["salary","left"])["left"].count().reset_index(name="count"))
salary_left_perc = salary_left.groupby(['salary', 'left']).agg({'count': 'sum'})
salary_left_perc = salary_left_perc.groupby(level=0).apply(lambda x:100 * x / float(x.sum()))
salary_left_perc.style.background_gradient(cmap = 'Greens')
```

The above table shows us the expected result which is that employees in the 'high' salary group have a much lower departure rate then (94% have stayed and 6% have left) 'low' salary group employees (79% stayed and 21% left). This  gives us the insight that the salary group the employee is in does impact their likeliness of leaving. 

```python
accident_left = pd.DataFrame(working_df.groupby(["work_accident","left"])["work_accident"].count().reset_index(name="count"))
accident_left_perc = accident_left.groupby(['work_accident', 'left']).agg({'count': 'sum'})
accident_left_perc = accident_left_perc.groupby(level=0).apply(lambda x:100 * x / float(x.sum()))
accident_left_perc.style.background_gradient(cmap = 'Greens')
```

An interesting insight is observed above, which is that the employees who had a work accident to not have a significantly higher tendancy to leave the company. While work accidents ideally need to be eliminated, we see that for departments we are investigating, work accidents are not a significant 'deal breaker' for employees.

```python
promotion_left = pd.DataFrame(working_df.groupby(["promotion_last_5years","left"])["left"].count().reset_index(name="count"))
promotion_left_perc = promotion_left.groupby(['promotion_last_5years', 'left']).agg({'count': 'sum'})
promotion_left_perc = promotion_left_perc.groupby(level=0).apply(lambda x:100 * x / float(x.sum()))
promotion_left_perc.style.background_gradient(cmap = 'Greens')
```

The employees of Accounting, HR and Technical department who have been promoted in the past 5 years had a lower departure rate than employees who haven't been promoted. The internal recognition show to be important for employees as we see that both the salary group they are in and whether or not they were promoted in the last 5 years have a positive correlation with their departure. 

Now let's look at the correlation matrix of the dataframe where we only have the data for departments Accounting, HR and Technical.

```python
corrMatrix2 = working_df.corr()
ax2 = sns.heatmap(round(corrMatrix2,2), annot=True, vmin=-1, vmax=1, linewidths=.5, cmap="Greens")
bottom, top = ax2.get_ylim()
ax2.set_ylim(bottom + 0.5, top - 0.5)
```

From the above correlation matrix we can see that there are only two noticable correlations (above 0.1) which are between employee departure ('left') and satisfaction level (negative correlation), and between employee departure ('left') and work accidents (negative correlation).

The negative correlation between work accident and departure was also discovered through our previous analysis above, where we learned that employees who had work accidents do not have a significantly higher tendancy to leave the company. 

On the other hand we see that the correlation between satisfaction level and departure are also negatively correlated in our dataframe where we only have the departments Accounting, HR and Technical. 

Even though there are no significant negative or positive correlations between other variables, we have observed from our exploratory analysis above that employees who have been promoted in the past 5 years of their experience in the company, and employees who are in the high/medium salary group are less likely to leave than the ones who haven't been promoted and are in the low salary group. As departure is negative correlated with satisfaction_level, lowering the departure rate is highly likely to make a positive impact on the satisfaction level amongst Accounting, HR and Technical department employees.

5. look further into 'Accounting' department as their employees are neither the least promoted or least paid group yet have the lowest satisfaction level in the whole organization as well as have one of the highest departure rates.
