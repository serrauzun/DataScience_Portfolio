```python
df = pd.read_csv('HMEQ_Loss_clean.csv')
```

```python
print(df.head())
```

Separate binary columns to different df to first drop then add

```python
df_binary = df[['TARGET_BAD_FLAG', 'REASON_DebtCon', 'REASON_HomeImp', 'REASON_Missing', 'JOB_Mgr', 'JOB_Missing', 'JOB_Office', 'JOB_Other', 'JOB_ProfExe', 'JOB_Sales', 'JOB_Self']]
```

```python
print(df_binary.head())
```

```python
x = df.copy()
x = x.drop(df_binary, axis=1 )
varNames = x.columns
```

```python
print(varNames)
```

```python
print(x.head())
```

```python
print(x.head())
print(x.describe())
print( "\n\n")
```

****** NORMALIZING THE DATA ******

```python
theScaler = MinMaxScaler()
theScaler.fit(x)
```

```python
x_MINMAX = theScaler.transform(x)
x_MINMAX = pd.DataFrame(x_MINMAX)
print(x_MINMAX.head())
print("\n\n")
```

```python
varNames_minmax = []
for i in varNames:    
    newName = "nor_" + i    
    varNames_minmax.append( newName )
print(varNames_minmax)
print( "\n\n")
```

```python
x_MINMAX.columns = varNames_minmax
print(x_MINMAX.head())
print( "\n\n")
```

```python
print(x_MINMAX.describe())
print("\n\n")
```

```python
x_new = pd.concat([x, x_MINMAX, df_binary], axis=1)
```

```python
x_new.head()
```

** STANDARDIZING THE DATA **

```python
theScaler = StandardScaler()
theScaler.fit(x)
```

```python
y_std = theScaler.transform(x)
y_std = pd.DataFrame(y_std)
print(y_std.head())
print( "\n\n")
```

```python
varNames_std = []
for i in varNames:    
    newName = "std_" + i    
    varNames_std.append( newName )
```

```python
y_std.columns = varNames_std
print(y_std.head())
print("\n\n")
```

```python
print(y_std.describe())
print("\n\n")
```

```python
y_new = pd.concat([x, y_std, df_binary], axis=1)
print(y_new.head())
```
